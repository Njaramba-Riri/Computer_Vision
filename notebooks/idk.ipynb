```python
import cv2
print("OpenCV:", cv2.__version__)
```

```python
cap= cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    cv2.imshow('video', img)

    if cv2.waitKey(33) == 27:
        cv2.imwrite("Images/image.jpeg", img) 
        break


    
cap.release()
cv2.destroyAllWindows()
```
